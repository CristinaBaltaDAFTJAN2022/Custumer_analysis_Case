```python
# Using seaborn library i have construct the distribution plots for the numerical variables
def distrib_plots(num_df):
    plt.figure(figsize =(12,10))
    for i, col in enumerate(num_df.columns, 1):
        plt.subplot(3,3,i)
        sns.histplot(num_df[col])
```
